```python
#Q1  Top 7 RICE PRODUCTION State Data(Bar_plot)

#1. Qustion 1 
import pandas as pd
import matplotlib.pyplot as plt


# Group by 'State Name' and sum the rice production
state_rice_production = df.groupby('State Name')['RICE PRODUCTION (1000 tons)'].sum()

# Select top 7 rice producing states
top7_states = state_rice_production.sort_values(ascending=False).head(7)

# Plotting the bar plot
plt.figure(figsize=(6,4))
plt.bar(top7_states.index, top7_states.values, color='orange')
plt.title('Top 7 Rice Production States')
plt.xlabel('State')
plt.ylabel('Rice Production (1000 tons)')
plt.xticks(rotation=90)
plt.show()
```

```python
#Q10 Sorghum Production (Kharif and Rabi) by Region

# 1. Total sorghum (Kharif + Rabi) per row
df['TOTAL SORGHUM (1000 tons)'] = (
    df['KHARIF SORGHUM PRODUCTION (1000 tons)'] +
    df['RABI SORGHUM PRODUCTION (1000 tons)']
)

# 2. Region mapping for the states you listed
north_states = ['Punjab', 'Haryana', 'Uttar Pradesh', 'Uttarakhand',
                'Himachal Pradesh']
south_states = ['Andhra Pradesh', 'Telangana', 'Karnataka',
                'Tamil Nadu', 'Kerala']
east_states  = ['West Bengal', 'Bihar', 'Jharkhand', 'Assam', 'Orissa']
west_states  = ['Maharashtra', 'Gujarat', 'Rajasthan']
central_states = ['Madhya Pradesh', 'Chhattisgarh']

def get_region(state):
    if state in north_states:
        return 'North'
    if state in south_states:
        return 'South'
    if state in east_states:
        return 'East'
    if state in west_states:
        return 'West'
    if state in central_states:
        return 'Central'
    return 'Other'

df['Region'] = df['State Name'].apply(get_region)

# 3. Group by Region and sum total sorghum
region_sorghum = df.groupby('Region')['TOTAL SORGHUM (1000 tons)'].sum()
region_sorghum = region_sorghum.sort_values(ascending=False)

# 4. Bar chart
plt.figure(figsize=(8,5))
bars = plt.bar(region_sorghum.index, region_sorghum.values, color='purple')

plt.title("Sorghum Production (Kharif + Rabi) by Region")
plt.xlabel("Region")
plt.ylabel("Sorghum Production (1000 tons)")

# Add values on top   #Used chat gpt
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'{height:,.0f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

```

```python
#Q12

# Group by state: total production and average yield
state_soy = df.groupby('State Name')[[
    'SOYABEAN PRODUCTION (1000 tons)',
    'SOYABEAN YIELD (Kg per ha)'
]].agg({
    'SOYABEAN PRODUCTION (1000 tons)': 'sum',
    'SOYABEAN YIELD (Kg per ha)': 'mean'
})

# Top 5 states by production
top5 = state_soy.sort_values(
    by='SOYABEAN PRODUCTION (1000 tons)',
    ascending=False
).head(5)


fig, ax1 = plt.subplots(figsize=(8,5))

# Bars: production
bars = ax1.bar(
    top5.index,
    top5['SOYABEAN PRODUCTION (1000 tons)'],
    color='saddlebrown'
)
ax1.set_ylabel('Production (1000 tons)', color='saddlebrown')
ax1.tick_params(axis='y', labelcolor='saddlebrown')
plt.xticks(rotation=45)

# Add numbers on bars
for b in bars:
    h = b.get_height()
    ax1.text(b.get_x() + b.get_width()/2, h,
             f'{h:,.0f}', ha='center', va='bottom', fontsize=8)

# Line: yield on second y-axis
ax2 = ax1.twinx()
ax2.plot(
    top5.index,
    top5['SOYABEAN YIELD (Kg per ha)'],
    marker='o',
    color='darkgreen'
)
ax2.set_ylabel('Yield (Kg per ha)', color='darkgreen')
ax2.tick_params(axis='y', labelcolor='darkgreen')

plt.title('Top 5 Soybean Producing States and Yield Efficiency')
ax1.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

```python
#Q15  Rice vs. Wheat Yield Across States


import numpy as np

# Average yield per state
state_yield = df.groupby('State Name')[[
    'RICE YIELD (Kg per ha)',
    'WHEAT YIELD (Kg per ha)'
]].mean()

# (Optional) limit to major states for readability
state_yield = state_yield.sort_index()   # or sort by one yield if you want

states = state_yield.index
x = np.arange(len(states))          # positions for states
width = 0.4                         # bar width

plt.figure(figsize=(10,5))

# Rice bars
plt.bar(x - width/2,
        state_yield['RICE YIELD (Kg per ha)'],
        width,
        label='Rice',
        color='orange')

# Wheat bars
plt.bar(x + width/2,
        state_yield['WHEAT YIELD (Kg per ha)'],
        width,
        label='Wheat',
        color='green')

plt.title("Rice vs Wheat Yield Across States")
plt.xlabel("State")
plt.ylabel("Yield (Kg per ha)")
plt.xticks(x, states, rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
